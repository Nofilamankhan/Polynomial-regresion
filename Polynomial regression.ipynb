# Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing dataset

```python
dataset= pd.read_csv("Position_Salaries.csv")
x = dataset.iloc[:,1:2].values
z = dataset.iloc[:,2].values

#Splitting dataset

from sklearn.model_selection import train_test_split
x_train, x_test, z_train, z_test=train_test_split(x,z,test_size = 0.2, random_state=0)

#Fitting linear regression

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,z)

#fitting Polynomial Regression

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
poly_reg.fit(x_poly,z)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,z)
```

# Visualising the Regression

```python
#Linear Regression

plt.scatter(x,z,color ='red')
plt.plot(x, lin_reg.predict(x), color = 'blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

#Polynomial Regression

plt.scatter(x,z,color ='red')
plt.plot(x, lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

#visualising polynomial regression for result(for higher resolution and smoother curve)

x_grid = np.arange(min(x),max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(x, z, color = 'red')
plt.plot(x_grid, lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Predicting 

```python
#Linear Regression

lin_reg.predict([[6.5]])

#Polynomial Regression

lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
